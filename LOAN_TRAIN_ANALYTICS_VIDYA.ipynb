## Gender

```python
gen_mode=loan_df['Gender'].mode()[0]
loan_df['Gender']=loan_df['Gender'].fillna(gen_mode)
```

## catogorical and numercial columns

```python
cat_cols=loan_df.select_dtypes(include='object').columns
val_cols=loan_df.select_dtypes(exclude='object').columns
```

## Marrired column

```python
mar_mode=loan_df['Married'].mode()[0]
loan_df['Married']=loan_df['Married'].fillna(mar_mode)
loan_df.isnull().sum()
```

## Dependents column

```python
dep_mode=loan_df['Dependents'].mode()[0]
loan_df['Dependents']=loan_df['Dependents'].fillna(dep_mode)
loan_df.isnull().sum()
```

## Self_Employed

```python
sel_mode=loan_df['Self_Employed'].mode()[0]
loan_df['Self_Employed']=loan_df['Self_Employed'].fillna(sel_mode)
loan_df.isnull().sum()
```

## LoanAmount

```python
lo_mode=loan_df['LoanAmount'].mode()[0]
loan_df['LoanAmount']=loan_df['LoanAmount'].fillna(lo_mode)
loan_df.isnull().sum()
```

## Loan_Amount_Term

```python

lo_mean=loan_df['Loan_Amount_Term'].mode().values[0]
loan_df['Loan_Amount_Term']=loan_df['Loan_Amount_Term'].fillna(lo_mean)
loan_df['Loan_Amount_Term']=loan_df['Loan_Amount_Term'].astype('object')


loan_df.isnull().sum()
```

## Credit_History`

```python
cred_mean=loan_df['Credit_History'].mode().values[0]
loan_df['Credit_History']=loan_df['Credit_History'].fillna(cred_mean)
loan_df['Credit_History']=loan_df['Credit_History'].astype('object')

loan_df.isnull().sum()
```

```python
for i in val_cols:
    plt.figure(figsize=(12, 8))
    
    # Plot before handling outliers
    plt.subplot(1, 2, 1)
    box1 = plt.boxplot(loan_df[i], patch_artist=True)
    plt.setp(box1["boxes"], facecolor="red")  # Box color
    plt.setp(box1["whiskers"], color="blue")  # Whisker color
    plt.setp(box1["caps"], color="red")       # Cap color
    plt.setp(box1["medians"], color="orange") # Median color
    plt.setp(box1["fliers"], markeredgecolor="blue")
    plt.title(f'{i}_BoxPlot')
    
    # Calculate IQR and bounds for capping
    q1 = np.percentile(loan_df[i], q=25)  # 1st Quartile
    q3 = np.percentile(loan_df[i], q=75)  # 3rd Quartile
    iqr = q3 - q1  # Interquartile Range
    lb = q1 - 1.5 * iqr  # Lower Bound
    ub = q3 + 1.5 * iqr  # Upper Bound

    # Capping outliers
    loan_df[i] = np.where(loan_df[i] < lb, lb, loan_df[i])
    loan_df[i] = np.where(loan_df[i] > ub, ub, loan_df[i])
    
    # Plot after handling outliers
    plt.subplot(1, 2, 2)
    box2 = plt.boxplot(loan_df[i], patch_artist=True)  # Enable color fill
    plt.setp(box2["boxes"], facecolor="lightgreen")    # Box color
    plt.setp(box2["whiskers"], color="darkgreen")      # Whisker color
    plt.setp(box2["caps"], color="brown")              # Cap color
    plt.setp(box2["medians"], color="yellow")          # Median color
    plt.setp(box2["fliers"], markeredgecolor="black")  # Outlier color
    plt.title('After_Outliers_Capping')
    
    plt.show()
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
dir(grid_tree)
```

```python
param_grid={'criterion':['gini','entropy'],
           'max_depth':[3,4,5,6,7,8],
           'min_samples_split':[2,3,4],
           'min_samples_leaf':[1,2,3,4],
           'random_state':[0,42]
           }
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.cv_results_
```

```python
grid_search.cv_results_.keys()
```
